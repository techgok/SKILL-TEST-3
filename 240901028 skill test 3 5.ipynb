```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# Data
transmitter = ['T1','T2','T1','T3','T2','T1','T3','T2','T1','T3']
strength = [75,50,90,40,85,70,30,60,80,20]
frequency = [2000,1500,2200,1300,2100,1800,1200,1600,1900,1100]
quality = ['Good','Bad','Good','Bad','Good','Good','Bad','Bad','Good','Bad']

# Encode transmitter
le = LabelEncoder()
transmitter_encoded = le.fit_transform(transmitter)

X = np.column_stack((transmitter_encoded, strength, frequency))

# Encode labels
y = np.array([1 if q=="Good" else 0 for q in quality])

model = DecisionTreeClassifier()
model.fit(X, y)

sample = [[le.transform(['T2'])[0], 80, 2000]]
pred = model.predict(sample)

print("Signal Quality:", "Good" if pred[0]==1 else "Bad")
```
