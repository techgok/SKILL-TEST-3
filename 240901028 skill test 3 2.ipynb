```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

X = np.array([
    [5,0.5],
    [5.5,0.4],
    [4.8,0.6],
    [6,0.3],
    [6.5,0.2],
    [6.2,0.25],
    [4.9,0.55],
    [5.8,0.35],
    [5.2,0.5],
    [6.3,0.22]
])

# Working=0, Defective=1
y = np.array([0,0,0,1,1,1,0,1,0,1])

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

sample = [[5.7,0.3]]
pred = model.predict(sample)

print("Status:", "Defective" if pred[0]==1 else "Working")
```
